```python
import numpy as np
import matplotlib.pyplot as plt
import itertools
```

## 자료형의 특수기능

###넘파이 특수기능

```python
#벡터화 연산
arr = np.array([1,2,3])
arr + 1
```

```python
#bool을 이용한 인덱싱
arr = np.array([1,2,3,4,5])
arr[arr>3]
```

```python
#넘파이 특화 메소드
arr = np.array([5,4,3,2,1])
arr.min(), arr.max(), arr.mean(), arr.argmin(), arr.argmax()
```

```python
# sort
arr = np.array([5,4,3,2,1])
arr.sort()
arr
```

### 리스트 특수기능

```python
# sort
lst = [5,4,3,2,1]
lst.sort()
lst
```

```python
# append
lst = [5,4,3,2,1]
lst.append(4)
lst
```

```python
# join
s = '-'
lst = ['x','y','z']
s.join(lst) # join(s,lst) 로 해석
```

### str 특수기능

```python
#format
s = "제 이름은 {}입니다."
name = "정민"
s.format(name)
```

```python
#capitalize
s = 'asdf'
s.capitalize()
```

## 슈도 알고리즘

```python
set([''.join(i) for i in itertools.permutations(['x','x','y','y','z'])])
```

```python
len(set([''.join(i) for i in itertools.permutations(['x','x','y','y','z'])]))
```

```python
# 컴프리헨션과 조건문
x = list(range(1,16))
f = lambda x: np.sin(np.pi/4*x)
sum([xi for xi in x if f(2+xi)*f(2-xi)<1/4])
```
